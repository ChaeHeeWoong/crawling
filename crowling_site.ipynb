```python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import os
import io

dir = os.getcwd()
url =''
id = ''
password = ''

driver = webdriver.Chrome('./chromedriver')
driver.get(url)

driver.implicitly_wait(3)
driver.find_element_by_xpath('//*[@name="userid"]').send_keys(id)
driver.implicitly_wait(1)
driver.find_element_by_xpath('//*[@name="password"]').send_keys(password)
driver.implicitly_wait(1)
driver.find_element_by_xpath('//*[@value="로그인"]').click()
driver.implicitly_wait(1)


title_data = []
contents_data = []


for page in range(200):

# 글 내용에 접근하기 위해 하이퍼링크 따오기
    time.sleep(3)
    driver.implicitly_wait(2)
    req = driver.page_source
    bs = BeautifulSoup(req, 'lxml')

    links_with_text = []

    driver.implicitly_wait(2)
    for a in bs.find_all('a',attrs={'class' : 'article'}, href=True):
        if a.text:
            links_with_text.append(a['href'])
    print (links_with_text)
    if (page % 10) == 0 :
        print (str(page * 20) + "개 크롤링 완료")

#크롤링하기
    for i in range(20):
        link = '' + links_with_text[i]
        driver.get(link)
        req = driver.page_source
        bs = BeautifulSoup(req,'html.parser')
        driver.implicitly_wait(2)
        contents = driver.find_element_by_xpath('//*[@id="container"]/div[2]/article/a/p').text.strip()
        title = driver.find_element_by_xpath('//*[@id="container"]/div[2]/article/a/h2').text.strip()

        time.sleep(1)
        f = io.open('title' + str(i)+'_'+str(page)+".txt", mode="w", encoding='utf-8')
        f.write(title)
        f.close()
        f = io.open('contents' + str(i) +'_'+ str(page)+".txt", mode="w", encoding='utf-8')
        f.write(contents)
        f.close()

        driver.implicitly_wait(3)
        driver.back() # 뒤로가기

    pa_num = page + 2
    pa_url = '' + str(pa_num)
    driver.get(pa_url)


    driver.quit()
    driver = webdriver.Chrome('./chromedriver')
    driver.implicitly_wait(3)

    driver.get(pa_url)

    driver.find_element_by_xpath('//*[@name="userid"]').send_keys(id)
    driver.implicitly_wait(1)
    driver.find_element_by_xpath('//*[@name="password"]').send_keys(password)
    driver.implicitly_wait(1)
    driver.find_element_by_xpath('//*[@value="로그인"]').click()
    driver.implicitly_wait(1)

```
